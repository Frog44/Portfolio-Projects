```python
# Setting the noise/redundant/missing data variables
ignored_cols  = ['School', 'School URL', 'District URL', 'City URL', 'Phone', 'Home Prices Ranking', 'Community Health Ranking']
```

```python
# Setting categorical cols
cat_cols = ['District',  'City', 'Zip', 'County',  'Is Title I', 'Is Charter', 'Is Magnet', 'Is Virtual',]
```

```python
# Setting the numerical cols
num_cols = ['Number Students', 'Number Full-time Teachers',
       'Student/Teacher Ratio', 'Percent Free/Disc Lunch',
       'Percent Homes Rented', 'Percent of Unemployment',
       'Percent Parents with Masters and above education',
       'Percent parents with less than High School education',
       'Percent Home Incomes with less than Median income']
```

```python
# Setting the dependent/target/label variable
y = 'Standard Score'
```

```python
# Importing the pycaret and setting the experiment
from pycaret.regression import *

regression_model_setup = setup(data,
                                   target  = y,
                                   ignore_features=ignored_cols,
                                   numeric_features=num_cols,
                                   categorical_features=cat_cols,
                                   train_size = 0.7)
```

```python
# Invoking ML algorithms
compare_models()
```

```python
#select best performing model
model = create_model('huber')
```

```python
# Predicting standard scores using the best model
# Loading the local dataset
local_data = pandas.read_csv('/content/2021 Texas High Schools Scores (1).csv')
local_data
```

```python
# Predicting standard scores using the best model
local_data_predictions = predict_model(model, local_data)
```
